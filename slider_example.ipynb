```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button
```

```python
# Initial parameters
t = np.linspace(0, 1, 1001)
amp = 2
freq = 3
```

```python
# Set up the plot
fig, ax = plt.subplots(figsize = (13, 8));
line = ax.plot(t, amp * np.sin(2*np.pi*freq*t), lw = 2);
ax.set_xlabel('Time (s)', fontsize = 16);
```

```python
# Make room for the sliders
fig.subplots_adjust(left = 0.25, bottom = 0.25)
```

```python
# Make the frequency slider
ax_freq = fig.add_axes([0.25, 0.1, 0.65, 0.03])
freq_slider = Slider(ax = ax_freq,
                     label = 'Frequency (Hz)',
                     valmin = 0.1,
                     valmax = 30,
                     valinit = 3)
```

```python
# Make an amplitude slider
ax_amp = fig.add_axes([0.1, 0.25, 0.0225, 0.63])
amp_slider = Slider(ax = ax_amp,
                    label = 'Amplitude',
                    valmin = 0,
                    valmax = 10,
                    valinit = 2,
                    orientation = 'vertical')
```

```python
# Function to update silder values
def update(val):
    line.set_ydata(amp_slider.val * np.sin(2*np.pi*freq_slider.val*t))
    fig.canvas.draw_idle()
```

```python
# Register the update function with the sliders
freq_slider.on_changed(update)
amp_slider.on_changed(update)
```

```python
plt.show()
```
